###1. GELOMBANG SINUS DAN COSINUS

Dalam kode ini, kita membuat empat sinyal yang berbeda dengan mengubah amplitudo dan fase dari masing-masing gelombang. Amplitudo menentukan tinggi gelombang, sedangkan fase mengatur posisi awalnya. Dengan memodifikasi kedua parameter ini, kita dapat melihat variasi bentuk sinyal yang dihasilkan. Grafik yang dihasilkan memudahkan untuk memahami bagaimana perubahan kecil pada faktor-faktor ini memengaruhi bentuk dan karakteristik gelombang secara keseluruhan.

```python
t = np.arange(0, 2, 0.0001)

# Membuat sinyal y1, y2, y3, y4 dengan modifikasi agar berbeda
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)
y2 = 1.5 * np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -0.8 * np.sin(2 * np.pi * 5 * t + np.pi/3)
y4 = 0.3 * np.cos(2 * np.pi * 6 * t + np.pi/2)

# Membuat plot
fig, ax = plt.subplots(4, 1, figsize=(12, 8))
ax[0].plot(t, y1)
ax[0].set_title("Sinyal Y1 (2 * sin(2π * 3 * t))")

ax[1].plot(t, y2)
ax[1].set_title("Sinyal Y2 (1.5 * cos(2π * 4 * t + π/4))")

ax[2].plot(t, y3)
ax[2].set_title("Sinyal Y3 (-0.8 * sin(2π * 5 * t + π/3))")

ax[3].plot(t, y4)
ax[3].set_title("Sinyal Y4 (0.3 * cos(2π * 6 * t + π/2))")

plt.tight_layout()
plt.show()
```

Simulasi Sinyal ECG

```python
import neurokit2 as nk
import matplotlib.pyplot as plt

# Parameter Sinyal
duration = 186
sampling_rate = 150
noise = 0.86
heart_rate = 80
random_state = 30408

# Simulasi sinyal ECG
ecg_signal = nk.ecg_simulate(duration=duration, sampling_rate=sampling_rate,
                              noise=noise, heart_rate=heart_rate,
                              random_state=random_state)

# Menampilkan sinyal ECG asli
plt.figure(figsize=(10, 4))
plt.plot(ecg_signal)
plt.title('ECG Sinyal Asli dengan Sampling Rate 150 Hz')
plt.xlabel('Waktu (sampel)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Downsampling

```python
from scipy import signal

# Fungsi untuk melakukan downsampling
def downsample(signal, original_rate, target_rate):
    factor = original_rate // target_rate
    return signal[::factor]

sampling_rates = [100, 50, 25, 10, 5]
plt.figure(figsize=(15, 10))

for i, fs in enumerate(sampling_rates, 1):
    downsampled_signal = downsample(ecg_signal, sampling_rate, fs)
    plt.subplot(len(sampling_rates), 1, i)
    plt.plot(downsampled_signal)
    plt.title(f'Sinyal Downsampled (fs = {fs} Hz)')
    plt.xlabel('Waktu (sampel)')
    plt.ylabel('Amplitudo')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

###Aliasing
Aliasing adalah fenomena yang muncul ketika sinyal diambil sampelnya dengan frekuensi yang tidak memadai. Jika laju sampling terlalu rendah, sinyal yang memiliki frekuensi lebih tinggi daripada frekuensi Nyquist akan tampak terdistorsi, seolah-olah memiliki frekuensi yang lebih rendah. Hal ini menciptakan artefak alias yang tidak ada dalam sinyal asli.

Contohnya, ketika kita mengurangi laju sampling menjadi 25 Hz, 10 Hz, atau 5 Hz, detail dalam sinyal ECG mulai hilang, dan bentuk gelombang mengalami distorsi yang jelas. Dengan kata lain, informasi yang terdapat dalam sinyal asli tampak tercampur dan tidak dapat dikenali.

###Kesimpulan
Proses downsampling dapat mengakibatkan kehilangan informasi dari sinyal asli. Semakin rendah laju sampling yang diterapkan, semakin sedikit detail yang terjaga, serta semakin besar risiko distorsi yang disebabkan oleh aliasing. Aliasing muncul ketika frekuensi dalam sinyal yang melebihi separuh laju sampling dipersepsikan sebagai frekuensi yang lebih rendah, yang mengakibatkan perubahan bentuk pada sinyal.

Melalui simulasi ini, kita dapat melihat bahwa ketika sinyal ECG disampling pada laju yang lebih rendah dari 50 Hz, bentuk sinyal mulai mengalami hilang dan distorsi. Ini menunjukkan bahwa penurunan laju sampling secara signifikan mengurangi informasi penting dari sinyal asli, yang dapat berakibat negatif pada analisis lebih lanjut.

###6. ORDER DALAM FILTERING
*   Filter Rendah (Low-Order Filter)
Filter rendah dengan order rendah memiliki desain yang sederhana dan memberikan respons frekuensi yang halus. Meskipun mereka efektif dalam menghilangkan noise pada frekuensi tinggi, filter ini mungkin kurang efisien dalam menahan frekuensi yang tidak diinginkan. Akibatnya, sinyal yang dihasilkan bisa mengalami efek ringing, di mana ada fluktuasi yang tidak diinginkan di sekitar transisi frekuensi.



*   Filter Tinggi (High-Order Filter)
Sebaliknya, filter tinggi dengan order tinggi memiliki kompleksitas yang lebih besar dan lebih mampu menahan frekuensi yang tidak diinginkan. Respons frekuensi mereka lebih tajam, memungkinkan pemisahan yang lebih jelas antara frekuensi yang diizinkan dan yang ditolak. Namun, filter ini dapat menyebabkan efek ringing yang lebih signifikan dan dapat menjadi lebih sensitif terhadap noise, yang dapat mempengaruhi kualitas sinyal keluaran.




```python
from scipy.signal import butter, lfilter

# Fungsi untuk membuat filter Butterworth
def butterworth_filter(data, cutoff, fs, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return lfilter(b, a, data)

# Parameter sinyal
fs = 150.0
T = 1.0
n = int(fs * T)
t = np.linspace(0, T, n, endpoint=False)

# Sinyal dengan noise
a = 0.02
f0 = 5
sinusoidal = np.sin(2 * np.pi * f0 * t)
noise = np.random.normal(0, 0.5, sinusoidal.shape)
data = sinusoidal + noise

# Frekuensi cutoff
cutoff = 10

# Daftar order yang akan diuji
orders = [2, 4, 6, 8]

# Plotting hasil
plt.figure(figsize=(15, 10))
plt.plot(t, data, label='Sinyal dengan Noise', alpha=0.5)

for order in orders:
    filtered_data = butterworth_filter(data, cutoff, fs, order)
    plt.plot(t, filtered_data, label=f'Filter Order {order}')

plt.title('Efek Berbagai Order pada Filter Butterworth')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

###7. BAND PASS

Kita akan membuat sinyal respirasi sintetis, menambahkan noise, dan kemudian menerapkan filter band-pass untuk menghilangkan noise dan menjaga komponen frekuensi yang relevan. Frekuensi Cutoff pada filter band-pass adalah batas frekuensi di mana sinyal akan diteruskan dan di luar batas tersebut, sinyal akan diblokir. Untuk sinyal respirasi, frekuensi yang relevan umumnya berada dalam rentang 0,1 Hz hingga 0,5 Hz, karena ini adalah kisaran frekuensi pernapasan yang normal untuk orang dewasa (sekitar 12-20 napas per menit). Oleh karena itu, kita akan menentukan frekuensi cutoff untuk filter band-pass berdasarkan kisaran ini, sehingga kita dapat memisahkan sinyal respirasi dari noise.

```python
# Fungsi untuk membuat filter band-pass
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    return lfilter(b, a, data)

# Parameter sinyal
fs = 1000
T = 5.0
n = int(T * fs)
t = np.linspace(0, T, n, endpoint=False)


f_breath = 0.25
respiration_signal = 0.5 * np.sin(2 * np.pi * f_breath * t)

# Tambahkan noise
noise = np.random.normal(0, 0.1, respiration_signal.shape)
data_with_noise = respiration_signal + noise

# Frekuensi cutoff untuk filter band-pass (0.1 Hz - 0.5 Hz)
lowcut = 0.1
highcut = 0.5

# Terapkan filter band-pass
filtered_signal = bandpass_filter(data_with_noise, lowcut, highcut, fs)

# Plot hasil
plt.figure(figsize=(15, 8))

# Plot sinyal asli dengan noise
plt.subplot(3, 1, 1)
plt.plot(t, data_with_noise, label='Sinyal dengan Noise', color='gray')
plt.title('Sinyal Respirasi dengan Noise')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.legend()

# Plot sinyal asli
plt.subplot(3, 1, 2)
plt.plot(t, respiration_signal, label='Sinyal Respirasi Asli', color='blue')
plt.title('Sinyal Respirasi Asli')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.legend()

# Plot sinyal setelah filter
plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal, label='Sinyal Setelah Filter', color='red')
plt.title('Sinyal Respirasi Setelah Filter Band-Pass')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.legend()

plt.tight_layout()
plt.show()
```

Hasil dari eksperimen ini menunjukkan filter band-pass dapat membersihkan sinyal respirasi dengan efektif, memisahkan komponen frekuensi yang relevan dari noise. Dengan memilih frekuensi cutoff yang sesuai, kita dapat memastikan bahwa informasi penting dari sinyal respirasi dipertahankan, sementara noise yang berada di luar rentang frekuensi yang relevan dapat dihilangkan.
